```python
df['Name']=df.Name.str.lower()
df['Nationality']=df.Nationality.str.lower()
df['Club']=df.Club.str.lower()
df['Preferred Foot']=df['Preferred Foot'].str.lower()
df['Position']=df.Position.str.lower()
df.duplicated().sum()
df.drop_duplicates().reset_index(drop=True)
```

Напишите функцию, которая добавит колонку с разбиением возраста по группам: до 20, от 20 до 30, от 30 до 36 и старше 36. Посчитайте количество футболистов в каждой категории.

```python
def group_age(age):
    if age < 20:
        return 'до 20'
    elif age in range(20, 30):
        return 'от 20 до 30'
    elif age in range(30, 36):
        return 'от 30 до 36'
    else:
        return 'старше 36'
    
df['Group_age'] = df['Age'].apply(group_age)
df
```

```python
df['Group_age'].value_counts()
```
